```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Datadataset 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

Problem Statement:
Set the job_posted_date column as the index of the DataFrame and display the first 5 rows.

Hint:
Use the set_index() function to set the index.

```python
df.set_index('job_posted_date', inplace=True)
df.head()
```

Problem Statement:
Reset the index of the DataFrame from the last problem (where job_posted_date column is the index) to the default integer index and drop the existing index. Display the first 5 rows.

Hint:
Use the reset_index() function to reset the index.

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

Problem Statement:
Create a pivot table to calculate the median salary_year_avg for each company_name and sort the index in descending order. Display the first 5 rows.

Hint:
Use the sort_index() method and ascending parameter to sort company_name in reverse order.

```python
pivot1 = df.pivot_table(values='salary_year_avg', index='company_name', aggfunc='median')
pivot1.sort_index(ascending=False)
```

```python
pivot1 = df.pivot_table(values='salary_year_avg', index='company_name', aggfunc='median')
pivot1.sort_index(inplace=True, ascending=False)
pivot1.head()
```

Problem Statement:
Create a pivot table to calculate the median salary_year_avg for each job_title_short, reset the index of the resulting pivot table, then set the index to salary_year_avg. Display the first 5 rows.

Hint:
Use the pivot_table() method to create the pivot table, use the reset_index() method to reset the index and set_index() method to set the index.

```python
pivot2 = df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
pivot2.reset_index(inplace=True)
pivot2.set_index('salary_year_avg', inplace=True)
pivot2.head()
```

```python
pivot_table = df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
pivot_table.reset_index(inplace=True)
pivot_table.set_index('salary_year_avg', inplace=True)
pivot_table.head()
```
