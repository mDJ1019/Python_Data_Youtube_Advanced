Problem Statement:
Create a new column job_posted_year that extracts the year from the job_posted_date column and display the first 5 rows.

```python
# Data Cleanup - mora se kolona koja se odnosi na datum konvertovati preko datetime metode da bi se koristili podaci o godinama, mesecima, satima itd.
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_posted_year'] = df.job_posted_date.dt.year

df.head(5)
```

Problem Statement:
Drop rows where the salary_year_avg column has missing values and display the first 5 rows of the cleaned DataFrame.

Problem Statement:
Group the DataFrame by job_title_short and calculate the average salary_hour_avg for each job title. Display the result.

```python
avg_salary_hour  = df.groupby('job_title_short')['salary_hour_avg'].mean()  # GROUPBY grupise po nazivu posla, posle grouby se u uglastim zagradama 
# navodi kolona za koju hocemo da izvrsimo neku operaciju, ovde srednju vrednost
avg_salary_hour
```

Problem Statement:
Find the job title with the lowest average hourly salary and display the job title and the salary.

```python
index_min = df.salary_hour_avg.idxmin()
row = df.iloc[index_min]
print(row.job_title_short, row.salary_hour_avg)  # MOJE, U CELIJI ISPOD JE POTPUNO!!!!
```

```python
min_hourly_index = df['salary_hour_avg'].idxmin()  #isto kao gore
min_hourly_job = df.iloc[min_hourly_index][['job_title_short', 'salary_hour_avg']] #  df.iloc[indeks] daje ceo red datafrejma, pa se u novim uglastim zagradama daje ili jedna kolona
# ili u listi vise kolona koje zelimo da prikazemo
min_hourly_job
```
