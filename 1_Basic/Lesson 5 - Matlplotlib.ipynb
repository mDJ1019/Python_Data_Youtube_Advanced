```python
plt.bar(job_counts.index, job_counts)
plt.title('Postings by Job Title')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45,  ha='right') # ha je horiznontal alignment, bez njega nije dobro poravnato
plt.show()
```
